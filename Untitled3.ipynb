```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
def fetch_dir(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if(i['job']=="Director"):
            L.append(i['name'])
            break
    return L
```

```python
df_mov['crew']=df_mov['crew'].apply(fetch_dir)
```

```python
df_mov['overview']=df_mov['overview'].apply(lambda x: x.split())
```

```python
df_mov['genres']=df_mov['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df_mov['keywords']=df_mov['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df_mov['cast']=df_mov['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df_mov['crew']=df_mov['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df_mov['tags']=df_mov['overview']+df_mov['genres']+df_mov['keywords']+df_mov['cast']+df_mov['crew']
```

```python
df=df_mov[['title','id','tags']]
```

```python
df
```

```python
df['tags']=df['tags'].apply(lambda x: " ".join(x))
```

```python
df
```

```python
df['tags']=df['tags'].apply(lambda x: x.lower())
```

```python
df
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
       y.append( ps.stem(i))
    return " ".join(y)
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words= 'english',max_features=5000)
```

```python
vector=cv.fit_transform(df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_distances
distance = cosine_distances(vector)
print(distance)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def reccomend(movie):
    movie_index=df[df['title']==movie].index[0]
    movies_list=sorted(list(enumerate(similarity[movie_index])),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(df.iloc[i[0]].title)
```
