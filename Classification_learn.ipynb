```python
import pandas as pd
```

```python
df = pd.read_csv("500hits.csv", encoding = 'latin-1')
```

```python
df
```

```python
df = df.drop(columns=['PLAYER','CS'])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df.iloc[:,0:13]
y = df.iloc[:,13]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=11,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn =  KNeighborsClassifier(n_neighbors=11)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
knn2 =  KNeighborsClassifier(n_neighbors=11, weights='distance')
```

```python
knn2.fit(X_train,y_train)
```

```python
y_pred2 = knn2.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred2))
```

```python
print(classification_report(y_test,y_pred2))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(ccp_alpha=0.04,criterion="entropy")
```

```python
tree.fit(X_train,y_train)
```

```python
y_pred3 = tree.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred3))
print(classification_report(y_test,y_pred3))
```

```python
features = pd.DataFrame(tree.feature_importances_,index = X.columns)
```

```python
features
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(
    n_estimators = 1000,
    criterion = "entropy",
    min_samples_split = 10,
    max_depth = 10,
    random_state = 42
)
```

```python
rf.fit(X_train,y_train)
```

```python
rf.score(X_test,y_test)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred_rf))
print(classification_report(y_test,y_pred_rf))
```
