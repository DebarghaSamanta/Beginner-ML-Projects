```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
```

```python
ohe1 = OneHotEncoder(handle_unknown='ignore',sparse_output = False).set_output(transform = 'pandas')
ohe2 = OneHotEncoder(handle_unknown='ignore',sparse_output = False).set_output(transform = 'pandas')
```

```python
df_train['Sex'] = df_train['Sex_female']
df_train = df_train.drop(columns = ['Sex_female','Sex_female'])
```

```python
df_test['Sex'] = df_test['Sex_female']
df_test = df_test.drop(columns = ['Sex_female','Sex_male'])
```

Sex part done

```python
df_train['Cabin'] = df_train['Cabin'].astype(str).str[0]
```

```python
df_train.groupby('Cabin')['Survived'].mean().sort_values(ascending=False)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(ccp_alpha = 0.04,criterion = 'entropy' )
known_train = df_train[df_train['Cabin'] != 'n']
unknown_train = df_train[df_train['Cabin'] == 'n']
features_cabin =['Pclass','Fare','Age','Family']
known_train = known_train.dropna(subset=features_cabin)
X_cabin = known_train[features_cabin]
y_cabin = known_train['Cabin']
```

```python
tree.fit(X_cabin,y_cabin)
unknown_X = unknown_train[features_cabin].fillna(unknown_train[features_cabin].median())  # fill missing numeric values
predicted_decks_train = tree.predict(unknown_X)
df_train.loc[unknown_train.index, 'Cabin'] = predicted_decks_train
```

```python
df_train.groupby('Cabin')['Survived'].mean().sort_values(ascending=False)
```

```python
cabin_encoding_label = ['D','B','E','C','A','F','G','T']
ordinal = OrdinalEncoder(categories = [cabin_encoding_label])
df_train['Cabin'] = ordinal.fit_transform(df_train[['Cabin']])
```

```python
df_train['Embarked'] = df_train['Embarked'].fillna(df_train['Embarked'].mode()[0])
embarked_cat = ['C','Q','S']
ordinal_embarked = OrdinalEncoder(categories=[embarked_cat])
df_train['Embarked'] = ordinal_embarked.fit_transform(df_train[['Embarked']])
```

```python
df_test.info()
```

Embarked encoding done

```python
df_train['Age'] = df_train['Age'].fillna(df_train['Age'].mean())
df_test['Age'] = df_test['Age'].fillna(df_test['Age'].mean())
```

```python
df_test.info()
```

```python
df_test['Fare'] = df_test['Fare'].fillna(df_test['Fare'].mean())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier,VotingClassifier,StackingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
features = ['Pclass','Sex','Age','Fare','Cabin','Embarked','Family']
X = df_train[features]
y = df_train['Survived']
X_test = df_test[features]
```

```python
X_train,X_valid,y_train,y_valid = train_test_split(X,y,random_state = 42, test_size = 0.2)
```

```python
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_valid_sc = scaler.transform(X_valid)
X_test_sc = scaler.transform(X_test)
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train_sc,y_train)
```

```python
y_pred_lr = lr.predict(X_valid_sc)
```

```python
print("Classification Report:\n", classification_report(y_valid, y_pred_lr))
print("Confusion Matrix:\n", confusion_matrix(y_valid, y_pred_lr))
```

```python
y_pred1 = lr.predict(X_test_sc)
```

```python
rf = RandomForestClassifier(random_state=42, class_weight = 'balanced')
rf.fit(X_train, y_train)
```

```python
y_pred_rf = rf.predict(X_valid)
```

```python
print( classification_report(y_valid, y_pred_rf))
```

```python
y_pred2 = rf.predict(X_test)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt', 'log2']
}

grid_search = GridSearchCV(RandomForestClassifier(random_state=42),
                           param_grid, 
                           cv=5, 
                           scoring='accuracy', 
                           n_jobs=-1)

grid_search.fit(X_train, y_train)
```

```python
y_pred_grid = grid_search.predict(X_valid)
```

```python
print( classification_report(y_valid, y_pred_grid))
```

```python
y_pred3 = grid_search.predict(X_test)
```

```python
gb = GradientBoostingClassifier(random_state=42)
gb.fit(X_train, y_train)

# Predictions
y_pred_gb = gb.predict(X_valid)

# Evaluation
print(classification_report(y_valid, y_pred_gb))
```

```python
y_pred4 = gb.predict(X_test)
```

```python
svc = SVC(C = 1.0, gamma ='scale',probability = True)
svc.fit(X_train_sc, y_train)

# Predictions
y_pred_svc = svc.predict(X_valid_sc)

# Evaluation
print(classification_report(y_valid, y_pred_svc))
```

```python
y_pred5 = svc.predict(X_test_sc)
```

```python
knn = KNeighborsClassifier(n_neighbors = 8, weights = 'uniform')
knn.fit(X_train_sc, y_train)

# Predictions
y_pred_knn = knn.predict(X_valid_sc)

# Evaluation
print(classification_report(y_valid, y_pred_knn))
```

```python
y_pred6 = knn.predict(X_test_sc)
```

```python
from xgboost import XGBClassifier

# Initialize the model
xgb = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

# Train
xgb.fit(X_train, y_train)

# Predict
y_pred_xgb = xgb.predict(X_valid)
print(classification_report(y_valid, y_pred_xgb))
```

```python
y_pred7 = xgb.predict(X_test)
```

```python
voting_clf = VotingClassifier(
    estimators=[
        ('lr', lr),
        ('rf', rf),
        ('grid', grid_search),
        ('gb', gb),
        ('knn', knn),
        ('svc', svc),
        ('xgb',xgb)
    ],
    voting='soft'
)

# Train
voting_clf.fit(X_train, y_train)
```

```python
y_pred_vc = voting_clf.predict(X_valid)
print(classification_report(y_valid, y_pred_vc))
```

```python
y_pred8 = voting_clf.predict(X_test)
```

```python
base_learners = [
        ('lr', lr),
        ('rf', rf),
        ('gb', gb),
        ('knn', knn),
        ('svc', svc),
        ('xgb',xgb)
]
meta_learner = grid_search
stacking_clf = StackingClassifier(
    estimators=base_learners,
    final_estimator=meta_learner,
    passthrough=False  # You can try True as well to include original features
)
```

```python
stacking_clf.fit(X_train, y_train)
y_pred_stc = stacking_clf.predict(X_valid)
print(classification_report(y_valid, y_pred_stc))
```

```python
y_pred9 = stacking_clf.predict(X_test)
```
