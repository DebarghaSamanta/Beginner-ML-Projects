```python
df = df.drop(columns=['track_id','track_genre','album_name','track_name'])
```

```python
df['explicit'] = df['explicit'].astype(int)
```

```python
df.head()
```

```python
df['artists'].nunique()
```

```python
artist_popularity = df.groupby('artists')['popularity'].mean()
```

```python
artist_rank = artist_popularity.rank(method='dense', ascending=True).astype(int)
```

```python
df['artist_encoded'] = df['artists'].map(artist_rank)
```

```python
df = df.drop(columns=['artists'])
```

```python
df.head()
```

```python
df.shape
```

```python
X = df.iloc[:,1:]
```

```python
y = df.iloc[:,0:1]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 42, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train =scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor (n_estimators=100,criterion='squared_error',max_depth = 5,min_samples_split = 5,ccp_alpha=0.05)
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
from sklearn.metrics import root_mean_squared_error,r2_score
```

```python
root_mean_squared_error(y_test,y_pred_rf)
```

```python
r2_score(y_test,y_pred_rf)
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(n_estimators=100, learning_rate=0.5, max_depth=5, random_state=42)
```

```python
xgb.fit(X_train,y_train)
```

```python
y_pred_xgb = rf.predict(X_test)
```

```python
r2_score(y_test,y_pred_xgb)
```
